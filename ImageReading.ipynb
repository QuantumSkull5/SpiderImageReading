```python
import cv2
import pytesseract
import numpy as np
from matplotlib import pyplot as plt
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\adamain\AppData\Local\Programs\Tesseract-OCR\tesseract.exe'


# imshow(String titleofimage, image, float size)
def imshow(title = "Image", image = None, size = 10):
    w= image.shape[0]
    h=image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
#discrepancies: 1.Terminal ID: (QALY, 2. Number : AXXXXXXXXXXXOO4 1, 3. Te
receipt = cv2.imread(r"C:\Users\adamain\Downloads\Receipt.jpg")
imshow("No Modifications", receipt, 10)
output_txt = pytesseract.image_to_string(receipt)
print("PyTesseract Extracted: {}".format(output_txt))
```

```python
from skimage.filters import threshold_local
# 1st trial using mean
# We get the Value component from the HSV color space. 
# then we apply adaptive thresholding to
Values = cv2.split(cv2.cvtColor(receipt, cv2.COLOR_BGR2HSV))[2] #gets the Value of every pixel in the image, stored in a 2D array
Threshold = threshold_local(Values, 25, offset=15, method="mean")

# Apply the threshold operation
thresh = (Values > Threshold).astype("uint8") * 255 #image to display with threshold editing done to it. changes all values to either black (0) if below threshold or white (255) if above threshold 
imshow("threshold_local", thresh)
output_txt = pytesseract.image_to_string(thresh)
print("PyTesseract Extracted: {}".format(output_txt))
```

```python
# 2nd trial using the gaussian method
# We get the Value component from the HSV color space
# then we apply adaptive thresholdingto
V = cv2.split(cv2.cvtColor(receipt, cv2.COLOR_BGR2HSV))[2]
T = threshold_local(V, 25, offset=15, method="gaussian")

# Apply the threshold operation
thresh = (V > T).astype("uint8") * 255
imshow("threshold_local", thresh)

output_txt = pytesseract.image_to_string(thresh)
print("PyTesseract Extracted: {}".format(output_txt))
```

```python
from matplotlib import pyplot as plt
from easyocr import Reader


#get results from reader
reader = Reader(['en'], gpu = False)
results = reader.readtext(thresh)
#change back from grayscale before editing color back onto it
if len(thresh.shape) == 2:
    thresh = cv2.cvtColor(thresh, cv2.COLOR_GRAY2BGR)

#edit the image
all_text = []
# iterate over our extracted text
for (bbox, text, prob) in results:
    # display the OCR'd text and the associated probability of it being text
    print(f" Probability of Text: {prob*100:.3f}% OCR'd Text: {text}")

    # get the bounding box coordinates
    (tl, tr, br, bl) = bbox
    tl = (int(tl[0]), int(tl[1]))
    tr = (int(tr[0]), int(tr[1]))
    br = (int(br[0]), int(br[1]))
    bl = (int(bl[0]), int(bl[1]))

    # Remove non-ASCII characters from the text so that
    # we can draw the box surrounding the text overlaid onto the original image
    text = "".join([c if ord(c) < 128 else "" for c in text]).strip()
    all_text.append(text)
    for i in text:
        if (i.lower() == "frame"):
            cv2.rectangle(thresh, tl, br, (255, 0, 0), 2)
            cv2.putText(thresh, text, (tl[0], tl[1]-10),
            cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)

# show the output image
imshow("OCR'd Image", thresh, size = 12)
print("EASY OCR Extracted: {}".format(all_text))
```
